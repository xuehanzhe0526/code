```python
# pandas读入
data = pd.read_csv(path)    # TV、Radio、Newspaper、Sales
x = data[['TV', 'Radio', 'Newspaper']]
# x = data[['TV', 'Radio']]
y = data['Sales']
```

```python
# 绘制1
plt.plot(data['TV'], y, 'ro', label='TV')
plt.plot(data['Radio'], y, 'g^', label='Radio')
plt.plot(data['Newspaper'], y, 'b*', label='Newspaer')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

```python
# 绘制2
plt.figure(figsize=(9,12),dpi=400) #dpi 影响像素
plt.subplot(311) # 3行 一列  第一个图
plt.plot(data['TV'], y, 'ro')
plt.title('TV')
plt.grid()
plt.subplot(312)
plt.plot(data['Radio'], y, 'g^')
plt.title('Radio')
plt.grid() # 背景格子
plt.subplot(313)
plt.plot(data['Newspaper'], y, 'b*')
plt.title('Newspaper')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state=1)
print x_train, y_train
print x_train.shape
print x_test.shape
```

```python
linreg = LinearRegression()
model = linreg.fit(x_train, y_train)
print model
print linreg.coef_ # 系数
print linreg.intercept_ # 截距
```

```python
y_pred = linreg.predict(x_test)
print y_pred
```

```python
print type(y_pred)
```

```python
mse = np.average((y_pred - y_test) ** 2)    # Mean Squared Error 均方误差
rmse = np.sqrt(mse)     # Root Mean Squared Error 
print mse, rmse
```

```python
t = np.arange(len(x_test))
plt.plot(t, y_test, 'r-', linewidth=2, label='Test')
plt.plot(t, y_pred, 'g-', linewidth=2, label='Predict')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```
